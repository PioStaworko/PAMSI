## Metodologia Crisp-DM
#### 1. Business Understanding (Zrozumienie biznesowe)
Cel projektu: Analiza danych struktury szerokopasmowej w USA pozyskanych w lutym 2023 roku, oraz wykorzystanie 3 algorytmów do wykonania predykcji na tych danych.


#### 2. Data understanding (Zrozumienie danych)
W celu lepszego zrozumienia danych ze strony fcc (Federal Communications Commission, odpowiedzialnej za projekt samknows) pobrano słownik wyjaśniający konkretne tabele oraz parametry w nich zawarte.

Przykład:
| **_curr_dlping.csv_** |          **_Latency under downstream load_**         |
|:---------------------:|:----------------------------------------------------:|
|        unit_id        |      Unique identifier for an   individual unit      |
|         dtime         |           Time test finished in local time           |
|         ddate         |           Date test finished in local time           |
|         target        |             Target hostname or IP address            |
|        rtt_avg        |              Average RTT in microseconds             |
|        rtt_min        |              Minimum RTT in microseconds             |
|        rtt_max        |              Maximum RTT in microseconds             |
|        rtt_std        | Standard Deviation in Measured   RTT in microseconds |
|       successes       |                  Number of successes                 |
|        failures       |                  Number of failures                  |


3. #### Data Preparation (Przygotowanie danych)
   Jest to etap czyszczenia danych, w przypadku jakichkolwiek brakach w danych, lub uszkodzonych danych.
   Na tym etapie odbywa się także selekcja danych, na których zostanie wykonane modelowanie. W naszym przypadku wybraliśmy dane:
   - bytes_sec   -> curr_lct_dl
   - rtt_avg     -> curr_udp_latency
   - rtt_std     -> curr_udp_latency
   - packets     -> curr_udpcloss 
   - jitter_up   -> curr_udp_jitter
   - jitter_down -> curr_udp_jitter
  
    Dane, jako że są 'rozrzucone' pomiędzy różnymi tabelami, należało połączyć. Zrobiliśmy to za pomocą unit_id czyli unikalnego id jednostki oraz czasu zebrania konkretnej danej (dtime).
